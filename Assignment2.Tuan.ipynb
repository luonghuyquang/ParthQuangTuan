```python
#import section
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
 #Show datas of 5 first rows
df = pd.read_csv('Country-data.csv')
df.head()
```

```python
df
```

```python
# Drop columns except income and gdpp
df.drop(['child_mort','country', 'exports', 'health', 'imports', 'inflation','life_expec','total_fer'],axis='columns',inplace=True)
df
```

```python
#Run K-means algorithm
km = KMeans(n_clusters=3)
yp = km.fit_predict(df)
yp
```

```python
#Add cluster to the table
df['cluster']= yp
df['cluster'].value_counts()
```

```python
#Cluster unique
df.cluster.unique()
```

```python
#Define each cluster 1 value
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
#Draw scatter plot
plt.scatter(df1['income'],df1['gdpp'],color='blue',label='0')
plt.scatter(df2['income'],df2['gdpp'],color='green',label='1')
plt.scatter(df3['income'],df3['gdpp'],color='red',label='2')
plt.xlabel('income')
plt.ylabel('gdpp')
plt.legend()
```

* Utilize the Elbow method

```python
sse = []  # Create an empty list to store SSE values for different k values
k_rng = range(1, 10)  # Define a range of k values from 1 to 9
for k in k_rng:
    km = KMeans(n_clusters=k)  # Create a KMeans model with the current k value
    km.fit(df)  # Fit the model to the data (assuming 'df' is your DataFrame)
    sse.append(km.inertia_)  # Append the SSE value to the list
```

```python
#Draw elbow graph
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

 Think how is the situation now different from supervised learning (kNN) case 

* KNN is a supervised learning algorithm mainly used for classification problems, whereas K-Means (aka K-means clustering) is an unsupervised learning algorithm.
* K in K-Means refers to the number of clusters, whereas K in KNN is the number of nearest neighbors (based on the chosen distance metric).
* K in KNN is determined by comparing the performance of algorithm using different values for K.
* There are few ways to determine the number of groups/clusters in a dataset prior to the K-means clustering, and that are:
    * Elbow Method,
    * Silhouette Score,
    * gap statistic.
